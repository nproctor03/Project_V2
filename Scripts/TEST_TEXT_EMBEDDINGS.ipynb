```python
x = ['5_O\'Clock_Shadow',
 'Arched_Eyebrows',
 'Attractive',
 'Bags_Under_Eyes',
 'Bald',
 'Bangs',
 'Big_Lips',
 'Big_Nose',
 'Black_Hair',
 'Blond_Hair',
 'Blurry',
 'Brown_Hair',
 'Bushy_Eyebrows',
 'Chubby',
 'Double_Chin',
 'Eyeglasses',
 'Goatee',
 'Gray_Hair',
 'Heavy_Makeup',
 'High_Cheekbones',
 'Male',
 'Mouth_Slightly_Open',
 'Mustache',
 'Narrow_Eyes',
 'No_Beard',
 'Oval_Face',
 'Pale_Skin',
 'Pointy_Nose',
 'Receding_Hairline',
 'Rosy_Cheeks',
 'Sideburns',
 'Smiling',
 'Straight_Hair',
 'Wavy_Hair',
 'Wearing_Earrings',
 'Wearing_Hat',
 'Wearing_Lipstick',
 'Wearing_Necklace',
 'Wearing_Necktie',
 'Young']

labels =[]

for label in x: 
    label = label.replace("_", " ")
    labels.append(label)
    
labels
```

```python
from os.path import exists
import numpy as np 
import torch
import clip
from PIL import Image
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-L/14", device=device, jit=True)   
```

```python
img_path = "C:\\Users\\noelp\\OneDrive\\Pictures\\000581.jpg"
#img_path = "C:\\Users\\noelp\\OneDrive\\Pictures\\000289.jpg"
image = Image.open(img_path)
```

```python
text = clip.tokenize(labels).to(device)
image_input = preprocess(image).unsqueeze(0).to(device)
```

```python
with torch.no_grad():
    image_features = model.encode_image(image_input)
    text_features = model.encode_text(text)
```

```python
image_features /= image_features.norm(dim=-1, keepdim=True)
text_features /= text_features.norm(dim=-1, keepdim=True)
similarity = (100.0 * image_features @ text_features.T).softmax(dim=-1)
values, indices = similarity[0].topk(5)
```

```python
print("\nTop predictions:\n")
for value, index in zip(values, indices):
    print(f"{labels[index]:>16s}: {100 * value.item():.2f}%")
```

```python
sentences=[]
for label in labels:
    sentence_1 = "A photo of a woman with "+label
    sentence_2 = "A photo of a woman wearing "+label
    sentence_3 = "A photo of a "+label+" woman"
    sentence_4 = "A photo of a man with "+label
    sentence_5 = "A photo of a man wearing "+label
    sentence_6 = "A photo of a man "+label+" man"
    
    sentences.append(sentence_1)
    sentences.append(sentence_2)
    sentences.append(sentence_3)
    sentences.append(sentence_4)
    sentences.append(sentence_5)
    sentences.append(sentence_6)

sentences.append("A photo of a young asain woman")
sentences.append("A photo of a young black man wearing sunglasses")
```
